```python
#Import the required modules
import pandas as pd
from collections import Counter
```

```python
# Let's read the dataframe and store it in 'df'
df=pd.read_csv('Naive-Bayes.txt',sep=',');
df
```

```python
#Gather the required details from the data frame

#Storing the number of features, we are excluding the last column because we are assuming it as Targrt Label.
features=len(df.columns)-1

#Storing the entire column of Targrt Label.
t=df.iloc[:,-1]

#Getting the unique value occurence of Targrt Label by using Counter
y=Counter(t)
y
```

```python
#Let's find out the Probabilities of each attribute for the given class i.e., P(X|Class)

#To store probabilites of each attribute I'm using a dictionary
probabilities={}

#Let's read the each feature and find the probability of each attribute for given class
for f in range(features-1):
    
    #let take the unique values of the feature
    feature=df.iloc[:,f].unique()
    #Store the entire column
    data=df.iloc[:,f]
    
    #Iterate the loop for each class of Targrt Label.
    for clas in y:
        
        #Iterate the loop for each attribute of feature.
        for attribute in feature:
            
            #We need count of the attribute for give class.
            count=0
            #I'm giving 'attribute-class' as the key name to my dictionary 'probabilities'.
            attribute_class=''
            
            #Now I want to find count of attribute in the data for the given class.
            for l in range(len(data)):
                #This the condition to count the no.of occurences of an attribute for the given class.
                if attribute==data[l] and clas==t[l] :
                    count+=1 
                    
            #The key name in the format of 'Attribute-Class', Ex: 'Young-Yes'
            attribute_class=str(attribute)+'-'+clas
            
            #I'm adding the 1 to the numerator and no.of distinct values feature to the denominator
            count=(count+1)/(y[clas]+len(feature))#Laplacian smoothing
            #Updating my dictionary
            
            probabilities[attribute_class]=count
probabilities
```

```python
#Till now I got the probabilities of all attributes for the given classes.

#Now lets apply the Naive Bayes Theorem to get the probable class for given attributes
def Naive(test):
    
    #I'm taking two lists to store the Output class and probability of outcme class
    outcome=[]
    outcomeProbability=[]
    
    #Lets find out the probability of test attributes for the given class
    for clas in y:
        
        #I need to find out the product of each attributes probability
        prod=1
        #Lets calculate the product of given attributes with our dictionary
        for attribute in test:
            #In my dictionary the attribute is in the form of 'Atteibute-class'. So, I need to modify attributes in such way.
            attribute_class=str(attribute)+'-'+clas
            #find products
            if attribute_class in probabilities:
                prod*=probabilities[attribute_class]

        outcome.append(clas)
        #We also need to multiply with the Probalility of class accroding to Naive Bayes Theorem
        prod*=y[clas]/len(t)
        outcomeProbability.append(prod)
        
    #Now I have the Probabilities of test attributes for given classes 
    #Next thing is to find out the maximum value of probabilities.
    #Which class had the Maximum probality, such class is the probable outcome for the given test attributes.
    print("The probable outcome for given attributes is : ",outcome[outcomeProbability.index(max(outcomeProbability))])
```

```python
#Give test attributes to the Naive function.
test=['Middle','Low','Yes','Good']
Naive(test)
```
